```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Create the countplot
ax = sns.countplot(data=data, x="SeniorCitizen", hue="Churn")

# Calculate counts for each combination of SeniorCitizen and Churn
counts = pd.crosstab(data['SeniorCitizen'], data['Churn'])

# Calculate percentages of Churn within each SeniorCitizen group
percentages = counts.apply(lambda r: r / r.sum() * 100, axis=1)

# Get the hue categories in the order they appear in the legend
hue_categories = [text.get_text() for text in ax.legend_.texts]

# Iterate through each container (which corresponds to a hue category)
for i, container in enumerate(ax.containers):
    current_churn_category = hue_categories[i]

    # Iterate through each bar in the container
    for bar_idx, bar in enumerate(container):
        height = bar.get_height()

        # Get the x-axis category label (SeniorCitizen) for the current bar
        # This assumes the order of x-tick labels matches the order of bars for each hue.
        senior_citizen_category = ax.get_xticklabels()[bar_idx].get_text()

        # Retrieve the pre-calculated percentage
        pct_value = percentages.loc[senior_citizen_category, current_churn_category]

        # Add the percentage label on top of the bar
        ax.annotate(f'{pct_value:.1f}%',
                    xy=(bar.get_x() + bar.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')

plt.title("SeniorCitizen vs Churn (Percentage)")
plt.show()
```

```python
senior_data = data[data['SeniorCitizen'] == 'yes']
churn_counts = senior_data['Churn'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(churn_counts, labels=churn_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Churn Distribution among Senior Citizens')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# List of columns to plot
cols = ['PhoneService', 'MultipleLines', 'InternetService',
        'OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport',
        'StreamingTV', 'StreamingMovies']

# Set up grid size (3 columns, enough rows to fit all plots)
n_cols = 3
n_rows = -(-len(cols) // n_cols)  # ceiling division

fig, axes = plt.subplots(n_rows, n_cols, figsize=(15, 4 * n_rows))
axes = axes.flatten()

for i, col in enumerate(cols):
    # Fixed: Removed the repeated 'hue' argument and set hue to 'Churn'
    sns.countplot(data=data, x=col, ax=axes[i], hue="Churn", palette='viridis')
    axes[i].set_title(col)
    axes[i].tick_params(axis='x', rotation=45)

# Hide any unused subplots
for j in range(len(cols), len(axes)):
    fig.delaxes(axes[j])

plt.tight_layout()
plt.show()
```
